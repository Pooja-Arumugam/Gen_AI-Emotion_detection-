```python
#Splitting data into train and test data
tr_text=df['sentence']
tr_label=df['label']

val_text=val_df['sentence']
val_label=val_df['label']

ts_text=ts_df['sentence']
ts_label=ts_df['label']

```

```python
encoder=LabelEncoder()
tr_label=encoder.fit_transform(tr_label)
val_label=encoder.transform(val_label)
ts_label=encoder.transform(ts_label)
```

```python
tokenizer=Tokenizer(num_words=10000)
tokenizer.fit_on_texts(tr_text)
sequences=tokenizer.texts_to_sequences(tr_text)
tr_x=pad_sequences(sequences,maxlen=50)
tr_y=to_categorical(tr_label)

sequences=tokenizer.texts_to_sequences(val_text)
val_x=pad_sequences(sequences,maxlen=50)
val_y=to_categorical(val_label)

sequences=tokenizer.texts_to_sequences(ts_text)
ts_x=pad_sequences(sequences,maxlen=50)
ts_y=to_categorical(ts_label)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, GlobalMaxPooling1D

# Define the model architecture
model = Sequential([
    Embedding(input_dim=10000, output_dim=128, input_length=50),  # Embedding layer
    LSTM(64, return_sequences=True),  # LSTM layer
    GlobalMaxPooling1D(),  # Pooling layer to reduce dimensions
    Dense(64, activation='relu'),  # Fully connected layer
    Dropout(0.5),  # Dropout to reduce overfitting
    Dense(4, activation='softmax')  # Output layer for multi-class classification
])

# Build the model by providing an input shape
model.build(input_shape=(None, 50))  # Batch size is None, sequence length is 50

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Display the model summary
model.summary()



```

```python
# Training the model
history = model.fit(
    tr_x, tr_y,  # Training data
    validation_data=(val_x, val_y),  # Validation data
    batch_size=64,  # Batch size
    epochs=10,  # Number of epochs
    verbose=1  # Display progress
)

# Evaluate the model on the test dataset
test_loss, test_accuracy = model.evaluate(ts_x, ts_y)
print(f"Test Accuracy: {test_accuracy:.4f}")

# Save the model for future use
model.save("emotion_classification_model.h5")
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load the saved model
model = load_model("emotion_classification_model.h5")

# Example sentences to predict
example_sentences = [
    "I am so happy and excited about my new job!",
    "I feel extremely sad and lonely today.",
    "I am really angry at how things turned out.",
    "I am afraid of what will happen tomorrow."
]

# Tokenize and pad the example sentences
example_sequences = tokenizer.texts_to_sequences(example_sentences)
example_padded = pad_sequences(example_sequences, maxlen=50)

# Perform prediction
predictions = model.predict(example_padded)

# Decode predictions
emotion_labels = encoder.classes_  # Get class labels from encoder
predicted_emotions = [emotion_labels[pred.argmax()] for pred in predictions]

# Display predictions
for sentence, emotion in zip(example_sentences, predicted_emotions):
    print(f"Sentence: '{sentence}' => Predicted Emotion: {emotion}")
```

```python
# Visualization
for i, sentence in enumerate(example_sentences):
    plt.figure(figsize=(8, 5))
    plt.bar(emotion_labels, predictions[i], color='green')
    plt.xlabel("Emotions")
    plt.ylabel("Predicted Probability")
    plt.title(f"Predicted Emotion Probabilities for:\n'{sentence}'")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7,color='red')
    plt.show()
```

```python
from transformers import pipeline

# Load a pre-trained generative model (e.g., GPT-2)
generator = pipeline("text-generation", model="gpt2")

# Example sentences for prediction
example_sentences = [
    "I am so happy and excited about my new job!",
    "I feel extremely sad and lonely today.",
    "I am really angry at how things turned out.",
    "I am afraid of what will happen tomorrow."
]

# Tokenize, pad, and classify emotions
example_sequences = tokenizer.texts_to_sequences(example_sentences)
example_padded = pad_sequences(example_sequences, maxlen=50)
predictions = model.predict(example_padded)
predicted_classes = np.argmax(predictions, axis=1)

# Generate explanations using generative AI
for i, sentence in enumerate(example_sentences):
    predicted_emotion = encoder.inverse_transform([predicted_classes[i]])[0]
    prompt = f"The sentence is classified as '{predicted_emotion}'. Generate a supportive explanation or insight for this emotion."

    # Use the generative AI to generate an explanation
    explanation = generator(prompt, max_length=50, num_return_sequences=1)[0]['generated_text']

    # Print results
    print(f"Sentence: {sentence}")
    print(f"Predicted Emotion: {predicted_emotion}")
    print(f"Generated Explanation:\n {explanation}")
    print("\n" + "-" * 50 + "\n")
```
